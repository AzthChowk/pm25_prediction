```python
# import necessary files and libraries
import pandas as pd 
```

```python
# list out the years and stations
years = [2022,2023,2024]
# stations = ['ratnapark','khumaltar','bhaktapur','kirtipur','shankhapark','pulchowk','bhaisipati']
stations = ['ratnapark','khumaltar']
```

```python
# Load the PM2.5 datasets and combine it by year and station-wise
try:
    yearly_pm25df = {}
    for j in stations:
        df_list = []
        for i in years:
            i_pm25df_j = pd.read_csv(f"./datasets/Pm25 record data/y_{i}_{j} PM2.5 hourlyaverage.csv")
            df_list.append(i_pm25df_j)
            print(f"{i}_pm25df_{j}")
        yearly_pm25df[j] = pd.concat(df_list, ignore_index = True)
        print(yearly_pm25df[f"{j}"])
        print("\n")
except:
    print("Something went wrong.")
```

```python
for i in stations:
    print(f"{i}".upper())
    yearly_pm25df[i]['date'] = pd.to_datetime(
        yearly_pm25df[i]['date'],
        format='mixed',
        errors='coerce'
    )

    # Now extract components
    yearly_pm25df[i]['year'] = yearly_pm25df[i]['date'].dt.year
    yearly_pm25df[i]['month'] = yearly_pm25df[i]['date'].dt.month
    yearly_pm25df[i]['day'] = yearly_pm25df[i]['date'].dt.day
    yearly_pm25df[i]['hour'] = yearly_pm25df[i]['date'].dt.hour
    print(yearly_pm25df[i].head())
```

```python
# -------------------------------------------------
# 1. Fill missing PM2.5 with month-hour average
# -------------------------------------------------
def fill_with_monthly_hourly_avg(df: pd.DataFrame) -> pd.Series:
    # Compute the average for every (year, month, hour) that actually exists
    monthly_hourly = (
        df.groupby(['year', 'month', 'hour'])['PM2.5']
          .mean()
          .rename('monthly_hourly_avg')
    )

    # Merge the averages back onto the original frame
    filled = df.merge(
        monthly_hourly,
        left_on=['year', 'month', 'hour'],
        right_index=True,
        how='left'
    )

    # Where PM2.5 is NaN â†’ use the monthly-hourly average
    return filled['PM2.5'].fillna(filled['monthly_hourly_avg'])

# -------------------------------------------------
# 2. Fill remaining NaNs with the global station average
# -------------------------------------------------
def fill_with_global_avg(df: pd.DataFrame) -> pd.Series:
    global_avg = df['PM2.5'].mean()
    return df['PM2.5'].fillna(global_avg)

# -------------------------------------------------
# Main loop over stations
# -------------------------------------------------
for station in stations:                     # <-- `stations` must be an iterable
    df = yearly_pm25df[station].copy()       # work on a copy to avoid side-effects

    # ---- show a preview (only in notebooks) ----
    try:
        display(f"--- Station: {station} ---")
        display(df.head())
    except NameError:
        print(f"\n--- Station: {station} ---")
        print(df.head())

    # ---- 1. month-hour imputation ----
    df['PM2.5'] = fill_with_monthly_hourly_avg(df)

    # (optional) inspect the monthly-hourly averages
    monthly_hourly = (
        df.groupby(['year', 'month', 'hour'])['PM2.5'].mean()
    )
    print("Monthly-hourly averages (sample):")
    print(monthly_hourly.head(12))

    # ---- 2. global fallback imputation ----
    overall_avg = df['PM2.5'].mean()
    print(f"Overall station average: {overall_avg:.3f}")

    df['PM2.5'] = fill_with_global_avg(df)

    # ---- store the cleaned frame back ----
    yearly_pm25df[station] = df

print("\nImputation finished for all stations.")
```

```python
metdf_list = []
for j in stations:
    metdf_j = pd.read_csv(f"./datasets/Meteorological parameters/metdata_{j}_2022to2024.csv")
    metdf_list.append(metdf_j)

for x in metdf_list:
        print(x)
```

```python
# merge the dataset
stations = list(yearly_pm25df.keys())  # ['ratnapark','khumaltar']

metdf_dict = {station: metdf_list[i] for i, station in enumerate(stations)}
combined = {}

for station in stations:
    pm = yearly_pm25df[station].reset_index(drop=True)
    met = metdf_dict[station].reset_index(drop=True)

    combined[station] = pd.concat([pm, met], axis=1)
    
```

```python
# Feature selection
for x in combined:
    combined[x] = combined[x][['PM2.5',	'YEAR',	'MO', 'DY', 'HR', 'PS', 'WS2M',	'WD2M',	'WS10M', 'WD10M', 'PRECTOTCORR','RH2M',	'QV2M',	'T2M']]
    display(combined[x].head())
```

```python
for x in combined:
    x_merged_df = combined[x]
    x_merged_df.to_csv(f"./datasets/merged_datasets/{x}_merged_df.csv")
```

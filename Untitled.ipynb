````python
# --- Import required libraries ---
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestRegressor, RandomForestClassifier
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score, accuracy_score
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
````

````python
# --- Step 1: Load the datasets ---
# Replace the file paths with your local ones if needed

pm_df = pd.read_csv('./datasets/Pm25 record data/y_2022_Ratnapark PM2.5 hourlyaverage.csv')
met_df = pd.read_csv('./datasets/Meteorological parameters/Ratnapark/POWER_Point_Hourly_20220101_20221231_027d70N_085d31E_LST.csv')

print("PM2.5 data:", pm_df.shape)
print("Meteorological data:", met_df.shape)
````

````python
met_df
````

````python
# --- Step 2: Preprocess PM2.5 dataset ---
# Ensure date column is parsed correctly
pm_df['date'] = pd.to_datetime(pm_df['date'], errors='coerce')

# Identify hour columns (excluding the Date column)
hour_cols = [col for col in pm_df.columns if col not in ['Date', 'date']]

# Compute daily peak value and corresponding hour
pm_df['peak_value'] = pm_df[hour_cols].max(axis=1)
pm_df['peak_hour'] = pm_df[hour_cols].idxmax(axis=1)

# Convert "Hour_15" → 15
pm_df['peak_hour'] = pm_df['peak_hour'].str.extract(r'(\d+)').astype(float)
````

````python
pm_df.to_csv('check.csv')
````

````python



# --- Step 3: Preprocess meteorological dataset ---
# Identify time or date column (POWER format often has 'YEAR', 'MO', 'DY', 'HR')
if {'YEAR', 'MO', 'DY'}.issubset(met_df.columns):
    met_df['Date'] = pd.to_datetime(met_df[['YEAR','MO','DY']])
else:
    # If single DATE/TIME column exists, parse it
    date_col = [c for c in met_df.columns if 'DATE' in c.upper() or 'TIME' in c.upper()][0]
    met_df['Date'] = pd.to_datetime(met_df[date_col])

# Aggregate hourly meteorological data into daily means (or sums for rain)
agg_funcs = {}
for col in met_df.columns:
    if col not in ['Date','YEAR','MO','DY','HR']:
        if 'PRECTOT' in col.upper() or 'RAIN' in col.upper():
            agg_funcs[col] = 'sum'
        else:
            agg_funcs[col] = 'mean'

met_daily = met_df.groupby(met_df['Date'].dt.date).agg(agg_funcs).reset_index()
met_daily.rename(columns={'index':'Date'}, inplace=True)
met_daily['Date'] = pd.to_datetime(met_daily['Date'])

# --- Step 4: Merge datasets by Date ---
df = pd.merge(pm_df[['Date','peak_value','peak_hour']], met_daily, on='Date', how='inner')

# Add temporal and calendar features
df['month'] = df['Date'].dt.month
df['dayofweek'] = df['Date'].dt.dayofweek
df['dayofyear'] = df['Date'].dt.dayofyear
df['prev_peak'] = df['peak_value'].shift(1).fillna(df['peak_value'].mean())

# Drop missing values if any
df = df.dropna()

print("Merged dataset shape:", df.shape)

# --- Step 5: Split data ---
features = [col for col in df.columns if col not in ['Date','peak_value','peak_hour']]
X = df[features]
y_value = df['peak_value']
y_hour = df['peak_hour']

X_train, X_test, y_train_val, y_test_val = train_test_split(X, y_value, test_size=0.2, random_state=42)
X_train_c, X_test_c, y_train_c, y_test_c = train_test_split(X, y_hour, test_size=0.2, random_state=42)

# --- Step 6: Train regression model (for peak concentration) ---
rf_reg = RandomForestRegressor(n_estimators=200, random_state=42)
rf_reg.fit(X_train, y_train_val)
y_pred_val = rf_reg.predict(X_test)

# --- Step 7: Evaluate regression model ---
mae = mean_absolute_error(y_test_val, y_pred_val)
rmse = np.sqrt(mean_squared_error(y_test_val, y_pred_val))
r2 = r2_score(y_test_val, y_pred_val)

print("\n=== Regression (Peak Concentration) ===")
print(f"MAE : {mae:.3f}")
print(f"RMSE: {rmse:.3f}")
print(f"R²  : {r2:.3f}")

# --- Step 8: Train classification model (for peak hour) ---
rf_clf = RandomForestClassifier(n_estimators=200, random_state=42)
rf_clf.fit(X_train_c, y_train_c)
y_pred_c = rf_clf.predict(X_test_c)

# --- Step 9: Evaluate classification model ---
acc = accuracy_score(y_test_c, y_pred_c)
print("\n=== Classification (Peak Hour) ===")
print(f"Accuracy: {acc:.3f}")

# Circular MAE for hours
def circular_mae(y_true, y_pred):
    diff = np.abs(y_true - y_pred)
    diff = np.where(diff > 12, 24 - diff, diff)
    return np.mean(diff)

cmae = circular_mae(y_test_c.values, y_pred_c)
print(f"Circular MAE (hours): {cmae:.3f}")

# --- Step 10: Feature importance visualization ---
importances = pd.Series(rf_reg.feature_importances_, index=X.columns).sort_values(ascending=False)
plt.figure(figsize=(10,6))
sns.barplot(x=importances.values[:15], y=importances.index[:15])
plt.title("Top 15 Important Features (Peak Concentration Model)")
plt.show()

# --- Step 11: Save results (optional) ---
df_results = pd.DataFrame({
    'Date': X_test.index,
    'Actual_Peak': y_test_val.values,
    'Predicted_Peak': y_pred_val
})
df_results.to_csv("predicted_peak_results.csv", index=False)

print("\nBaseline models trained and evaluated successfully.")
```
````
